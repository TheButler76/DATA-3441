```python
import pandas as pd
import matplotlib.pyplot as plt
import math

data = pd.read_csv('EX02-144MEIS.csv')
```

```python
class Stats:
    def __init__(self, data):
        self.data = data

    #====================*
    # Summation for Mean |
    #====================*
    def recursive_sum(self, n=None):
        if n==None:
            n = len(self.data)
        if n==0:
            return 0
        else:
            return self.data[n-1] + self.recursive_sum(n-1)

    #======*
    # Mean |
    #======*
    def recursive_mean(self):
        summation = self.recursive_sum()
        n = len(self.data)
        if n==0:
            return ValueError("N is zero.")
        else:
            return summation / n



    #====================*
    # Standard Deviation |
    #====================*
    def recursive_squared_deviation_sum(self, mean_val=None, n=None):
        if mean_val is None:
            mean_val = self.recursive_mean()
        if n is None:
            n = len(self.data)
        if n == 0:
            return 0
        else:
            deviation = self.data[n - 1] - mean_val
            return deviation**2 + self.recursive_squared_deviation_sum(mean_val, n - 1)

    def standard_deviation(self):
        n = len(self.data)
        if n < 2:
            return None  # Not enough data for sample std dev
        squared_sum = self.recursive_squared_deviation_sum()
        return (squared_sum / (n - 1)) ** 0.5
    
        
#==================*
# Basic Statistics |
#==================*
x_stats = Stats(data['DwellPermit'])
x_SUM = x_stats.recursive_sum()
x_MEAN = x_stats.recursive_mean()
print("x mean: ", x_MEAN)
x_STD = x_stats.standard_deviation()
print("x std: ", x_STD)

y_stats = Stats(data['Production'])
y_SUM = y_stats.recursive_sum()
y_MEAN = y_stats.recursive_mean()
print("y mean: ", y_MEAN)
y_STD = y_stats.standard_deviation()
print("y std: ", y_STD)


#==========*
# Z Scores |
#==========*
def z_scores(data, mean, std):
    z_scores = []
    for data_point in data:
        z_score = ((data_point - mean) / std)
        z_scores.append(z_score)
    return z_scores
z_scores_x = z_scores(data['DwellPermit'], x_MEAN, x_STD)
z_scores_y = z_scores(data['Production'], y_MEAN, y_STD)


#=============*
# Pearson's r |
#=============*
def correlation(zx, zy):
    n = len(zx)
    summation = sum([zx[i] * zy[i] for i in range(n)])
    r = summation / (n-1)
    return r
R = correlation(z_scores_x, z_scores_y)
print("Pearson's r: ", R)
#==============================*
# Coefficient of determination |
#==============================*
print(f"Pearson's r squared: {R**2}")

#===================*
# slope of the line |
#===================*
def b_1(std_x, std_y, r):
    if r==None:
        return False
    else:
        return r * (std_y / std_x)
b1 = b_1(x_STD, y_STD, R)
print("b1: ", b1)

#===========*
# intercept |
#===========*
def b_0(y_mean, x_mean, b1):
    if b1==None:
        return False
    else:
        return y_mean - (b1 * x_mean)
b0 = b_0(y_MEAN, x_MEAN, b1)
print("b0: ", b0)

#=================*
# Predicted value |
#=================*
def y_hat(x, b0, b1):
    return b0 + (b1 * x)

def plotting():
    #===========================*
    # Plotting using matplotlib |
    #===========================*
    x_vals = data['Production'] 
    y_vals = data['DwellPermit']
    
    # Predicted line
    x_line = x_vals
    y_line = [y_hat(x, b0, b1) for x in x_line]
    
    # Plot
    plt.scatter(x_vals, y_vals, color='blue', label='Actual Data')
    plt.plot(x_line, y_line, color='red', label='Regression Line')
    plt.xlabel('Dwell Permit')
    plt.ylabel('Production')
    plt.title('Least Squares Regression Line')
    plt.legend()
    plt.grid(True)
    plt.show()
plotting()
```
